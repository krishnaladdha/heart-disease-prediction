<h2 style='color:darkvoilet' align="center">Let's Prepare the Model:</h2>

```python
X = dataset.drop('target',axis='columns')
y = dataset.target
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

```python
models = [LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier()]
scores = []
model_names = []
for model in models:
    scores.append(np.mean(cross_val_score(model, X_train, y_train, n_jobs=3, verbose=2, cv=5)))
    model_names.append(model.__class__.__name__)
```

```python
scores
```

```python
plt.figure(figsize=(10,10))
g=sns.barplot(model_names, scores)
ax=g
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11,  xytext=(0, 10),
                 textcoords='offset points')
```

<h2 style='color:royalblue' align="center">Lets use the Logistic Regression:</h2>

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
#Lets draw a Heat Map to Visualize:
sns.heatmap(cm, annot=True)
```

```python
#Prepare Classification Report:
report = classification_report(y_test,y_pred)
```
